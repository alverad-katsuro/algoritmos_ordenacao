```python
def quickSort(lista, show_output=True):
  before = datetime.now()
  instrucoes, comparacoes, trocas = quickSortModulo(lista, 0, len(lista) - 1)
  after = datetime.now()
  if show_output:
    print(f"Tempo para ordenar {(after - before).total_seconds()}\Instrucções {instrucoes}\nLista de saida -> {lista}")
  return len(lista), instrucoes, (after - before).total_seconds(), comparacoes, trocas

def quickSortModulo(lista, primeiro, ultimo, instrucoes=0, comparacoes=0, trocas=0):
  if primeiro<ultimo:
    indice_pivo, instrucoes, comparacoes, trocas = partition(lista, primeiro, ultimo, instrucoes, comparacoes, trocas)
    instrucoes, comparacoes, trocas = quickSortModulo(lista, primeiro, indice_pivo-1, instrucoes, comparacoes, trocas)
    instrucoes, comparacoes, trocas = quickSortModulo(lista, indice_pivo+1, ultimo, instrucoes, comparacoes, trocas)
    instrucoes += 4
    comparacoes += 1
  return instrucoes, comparacoes, trocas

def partition(lista, primeiro, ultimo, instrucoes, comparacoes, trocas):
  pivo = lista[randint(0, len(lista) - 1)]
  esquerda = primeiro+1
  direita = ultimo
  instrucoes += 3
  while esquerda <= direita:
    while (esquerda <= direita and lista[esquerda] <= pivo):
      esquerda = esquerda + 1
      instrucoes += 3
      comparacoes += 2

    while (lista[direita] >= pivo and direita >= esquerda):
      direita = direita -1
      instrucoes += 3
      comparacoes += 2

    if (direita > esquerda):
      comparacoes += 1
      lista[esquerda], lista[direita] = lista[direita], lista[esquerda]
      instrucoes += 2
      trocas += 1

  lista[primeiro], lista[direita] = lista[direita], lista[primeiro]
  trocas += 1
  instrucoes += 1
  return direita, instrucoes, comparacoes, trocas
```

```python
def shellSort(lista, incremento=None):
    before = datetime.now()
    instrucoes = 2
    sequencia_inc = []
    comparacoes = 0
    trocas = 0
    if incremento == None:
        incremento = len(lista) // 2
    while (incremento > 0):
      comparacoes += 1
      for startposition in range(incremento):
        instrucoes += 1
        temp = gapInsertionSort(lista, startposition, incremento)
        instrucoes += temp[0]
        comparacoes += temp[1]
        trocas += temp[2]

      sequencia_inc.append(incremento)

      incremento = incremento // 2
      instrucoes += 1
    after = datetime.now()
    return len(lista), instrucoes, (after - before).total_seconds(), comparacoes, trocas


def gapInsertionSort(lista,start,gap, instrucoes=0, comparacoes=0, trocas=0):
    for i in range(start+gap,len(lista),gap):

        currentvalue = lista[i]
        position = i
        instrucoes += 3
        while position>=gap and lista[position-gap]>currentvalue:
            comparacoes += 2
            trocas += 1
            instrucoes += 3
            lista[position]=lista[position-gap]
            position = position-gap
        
        instrucoes += 1
        lista[position]=currentvalue
    return instrucoes, comparacoes, trocas
```

```python
def heapify(lista, n, pai, instrucoes=0, comparacoes=0, trocas=0):
    maior_valor = pai
    filho_esquerda = 2 * pai + 1
    filho_direita = 2 * pai + 2
    instrucoes += 3
    if filho_esquerda < n and lista[pai] < lista[filho_esquerda]:
        instrucoes += 3
        comparacoes += 2
        maior_valor = filho_esquerda
    if filho_direita < n and lista[maior_valor] < lista[filho_direita]:
        instrucoes += 3
        comparacoes += 2
        maior_valor = filho_direita
    if maior_valor != pai:
        instrucoes += 2
        comparacoes += 1
        trocas +=1
        (lista[pai], lista[maior_valor]) = (lista[maior_valor], lista[pai])  # swap
        temp = heapify(lista, n, maior_valor)
        if temp != None:
            instrucoes += temp[0]
            comparacoes += temp[1]
            trocas += temp[2]
        return instrucoes, comparacoes, trocas
  
def heapSort(lista):
    n = len(lista)
    before = datetime.now()
    instrucoes = 1
    comparacoes = 0
    trocas = 0
    
    for i in range(n // 2 - 1, -1, -1):
        temp = heapify(lista, n, i)
        if temp != None:
            instrucoes += temp[0]
            comparacoes += temp[1]
            trocas += temp[2]

    for i in range(n - 1, 0, -1):
        (lista[i], lista[0]) = (lista[0], lista[i])
        instrucoes += 1
        temp = heapify(lista, i, 0)
        if temp != None:
            instrucoes += temp[0]
            comparacoes += temp[1]
            trocas += temp[2]
        
    after = datetime.now()
    return len(lista), instrucoes, (after - before).total_seconds(), comparacoes, trocas
```

```python
def grafico(dados, title):
  colunas = ["Tamanho", "Instruções", "Tempo (s)", "Comparacoes", "Trocas"]
  df = pd.DataFrame(dados, columns=colunas)
  grafi, axes = plt.subplots(nrows = 1, ncols=4, figsize=(12, 4), tight_layout=True)
  grafi.suptitle(title, fontsize=18, fontweight='bold')
  
  for i in range(len(axes)):
    axes[i].plot(df[colunas[0]], df[colunas[i + 1]], color='red', linestyle='solid', markerfacecolor='blue', markersize=5) #, marker='o' - criar pontos
    axes[i].set_xlabel(colunas[0], fontsize='large', fontweight='bold')
    axes[i].set_ylabel(colunas[i+1], fontsize='large', fontweight='bold')
    axes[i].axhline(color="black")
    axes[i].axvline(color="black")
    axes[i].tick_params(axis='x', rotation=60)

  plt.show()
```

```python
def compara_tempo(a, b, c, legenda=('QuickSort', 'HeapSort', 'ShellSort')):
  plt.title("Tempo")
  plt.plot(a["Tamanho"], a["Instruções"])
  plt.plot(b["Tamanho"], b["Instruções"])
  plt.plot(c["Tamanho"], c["Instruções"])
  plt.legend(legenda, loc='upper left')
  plt.xlabel("Tamanho")
  plt.ylabel("Instruções")
  plt.show()
```

```python
def compara_troca(a, b, c, legenda=('QuickSort', 'HeapSort', 'ShellSort')):
  plt.title("Trocas")
  plt.plot(a["Tamanho"], a["Trocas"])
  plt.plot(b["Tamanho"], b["Trocas"])
  plt.plot(c["Tamanho"], c["Trocas"])
  plt.legend(legenda, loc='upper left')
  plt.xlabel("Tamanho")
  plt.ylabel("Trocas")
  plt.show()
```

```python
def compara_comparacoes(a, b, c, legenda=('QuickSort', 'HeapSort', 'ShellSort')):
  plt.title("Comparações")
  plt.plot(a["Tamanho"], a["Comparacoes"])
  plt.plot(b["Tamanho"], b["Comparacoes"])
  plt.plot(c["Tamanho"], c["Comparacoes"])
  plt.legend(legenda, loc='upper left')
  plt.xlabel("Tamanho")
  plt.ylabel("Comparacoes")
  plt.show()
```

```python
def dados_quic_0_100k(array_random):
  dados_quickSort = []
  for a in array_random:
        tamanho = 0
        instrucao = 0
        tempo = 0
        comparacoes = 0
        trocas = 0
        for _ in range(3):
            lista = quickSort(a.copy(), show_output=False)
            tamanho += lista[0]
            instrucao += lista[1]
            tempo += lista[2]
            comparacoes += lista[3]
            trocas += lista[4]
        dados_quickSort.append([tamanho / 3, instrucao / 3, tempo / 3, comparacoes / 3, trocas / 3])
  return dados_quickSort
```

```python
def dados_heap_0_100k(array_random):
  dados_heap = []
  for a in array_random:
        tamanho = 0
        instrucao = 0
        tempo = 0
        comparacoes = 0
        trocas = 0
        for _ in range(3):
            lista = heapSort(a)
            tamanho += lista[0]
            instrucao += lista[1]
            tempo += lista[2]
            comparacoes += lista[3]
            trocas += lista[4]
        dados_heap.append([tamanho / 3, instrucao / 3, tempo / 3, comparacoes / 3, trocas / 3])
  return dados_heap
```

```python
def dados_shell_0_100k(array_random):
  dados_shell = []
  for a in array_random:
        tamanho = 0
        instrucao = 0
        tempo = 0
        comparacoes = 0
        trocas = 0
        for _ in range(3):
            lista = shellSort(a)
            tamanho += lista[0]
            instrucao += lista[1]
            tempo += lista[2]
            comparacoes += lista[3]
            trocas += lista[4]
        dados_shell.append([tamanho / 3, instrucao / 3, tempo / 3, comparacoes / 3, trocas / 3])
  return dados_shell
```

```python
def teste(dados, title, legenda=("Quicksort", "Heapsort", "Shellsort")):
  colunas = ["Tamanho", "Instruções", "Tempo (s)", "Comparacoes", "Trocas"]
  grafi, axes = plt.subplots(nrows = 1, ncols=3, figsize=(20, 6), tight_layout=True)
  grafi.suptitle(title, fontsize=20, fontweight='bold')
  
  for i in range(len(axes)):
    axes[i].plot(dados[0][colunas[0]], dados[0][colunas[i + 1]], color='red', linestyle='solid', markerfacecolor='blue', markersize=5)
    axes[i].plot(dados[1][colunas[0]], dados[1][colunas[i + 1]], color='blue', linestyle='solid', markerfacecolor='blue', markersize=5)
    axes[i].plot(dados[2][colunas[0]], dados[2][colunas[i + 1]], color='green', linestyle='solid', markerfacecolor='blue', markersize=5)
    axes[i].set_xlabel(colunas[0], fontsize=18, fontweight='bold')
    axes[i].set_ylabel(colunas[i+1], fontsize=18, fontweight='bold')
    axes[i].tick_params(axis='x', rotation=60, labelsize=13)
    axes[i].legend(("Quicksort", "Heapsort", "Shellsort"), loc='upper left', prop={'size': 15})
  plt.savefig(f"{title}.png")
  plt.show()
```
