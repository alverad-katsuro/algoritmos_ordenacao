```python
from random import sample
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt
```

```python
from random import randint
```

```python
dez = sample(range(200_000, 1, -2),  10)
cem = sample(range(1, 200_000),  100)
mil = sample(range(1, 200_000),  1_000)
dezmil = sample(range(1, 200_000),  10_000)
cemmil = sample(range(1, 200_000),  100_000)
```

```python
def quickSort(lista, show_output=True):
  before = datetime.now()
  comparacoes = 0
  trocas = 0
  comparacoes, trocas = quickSortModulo(lista, 0,len(lista)-1, comparacoes, trocas)
  after = datetime.now()
  if show_output:
    print(f"Tempo para ordenar {(after - before).total_seconds()}\nComparacoes {comparacoes}\nTrocas {trocas}\nLista de saida -> {lista}")
  return len(lista), comparacoes, trocas, (after - before).total_seconds()

def quickSortModulo(lista, primeiro, ultimo, comparacoes, trocas):
  if primeiro<ultimo:
    indice_pivo, comparacoes, trocas = partition(lista, primeiro, ultimo, comparacoes, trocas)
    comparacoes, trocas = quickSortModulo(lista, primeiro, indice_pivo-1, comparacoes, trocas)
    comparacoes, trocas = quickSortModulo(lista, indice_pivo+1, ultimo, comparacoes, trocas)
  return comparacoes, trocas

def partition(lista, primeiro, ultimo, comparacoes, trocas):
  pivo = lista[randint(0, len(lista) - 1)]
  esquerda = primeiro+1
  direita = ultimo
  while esquerda <= direita:
    while (esquerda <= direita and lista[esquerda] <= pivo):
      esquerda = esquerda + 1
      comparacoes +=1

    while (lista[direita] >= pivo and direita >= esquerda):
      direita = direita -1
      comparacoes +=1

    if (direita > esquerda):
      lista[esquerda], lista[direita] = lista[direita], lista[esquerda]
      trocas += 1
      comparacoes +=1

  lista[primeiro], lista[direita] = lista[direita], lista[primeiro]
  trocas += 1
  return direita, comparacoes, trocas
```

```python
def shellSort(lista, incremento=None):
    before = datetime.now()
    comparacoes = 0
    trocas = 0
    sequencia_inc = []
    if incremento == None:
        incremento = len(lista) // 2
    while (incremento > 0):
      for startposition in range(incremento):
        troc_t, comp_t = gapInsertionSort(lista,startposition,incremento, trocas, comparacoes)
        trocas += troc_t
        comparacoes += comp_t

      sequencia_inc.append(incremento)

      incremento = incremento // 2
    after = datetime.now()
    return len(lista), comparacoes, trocas, (after - before).total_seconds()


def gapInsertionSort(lista,start,gap, trocas, comparacoes) -> list[int]:
    for i in range(start+gap,len(lista),gap):

        currentvalue = lista[i]
        position = i
        while position>=gap and lista[position-gap]>currentvalue:
            comparacoes += 1
            trocas += 1
            lista[position]=lista[position-gap]
            position = position-gap
        
        trocas += 1
        lista[position]=currentvalue
    return [trocas, comparacoes]
```

```python
def heapify(lista, n, pai, comparacoes=0, trocas=0):
    maior_valor = pai  # Initialize largest as root
    filho_esquerda = 2 * pai + 1  # left = 2*i + 1
    filho_direita = 2 * pai + 2  # right = 2*i + 2
 
 # See if left child of root exists and is
 # greater than root
    if filho_esquerda < n and lista[pai] < lista[filho_esquerda]:
        comparacoes += 1
        trocas += 1
        maior_valor = filho_esquerda
 
 # See if right child of root exists and is
 # greater than root
 
    if filho_direita < n and lista[maior_valor] < lista[filho_direita]:
        comparacoes += 1
        trocas += 1
        maior_valor = filho_direita
 
 # Change root, if needed
 
    if maior_valor != pai:
        trocas += 1
        (lista[pai], lista[maior_valor]) = (lista[maior_valor], lista[pai])  # swap
 
  # Heapify the root.
        heapify(lista, n, maior_valor, comparacoes, trocas)
        return comparacoes, trocas
 
 
# The main function to sort an array of given size
 
def heapSort(lista):
    n = len(lista)
    before = datetime.now()
    comparacoes = 0
    trocas = 0
    
    for i in range(n // 2 - 1, -1, -1):
        lista_temp = heapify(lista, n, i)
        if lista_temp != None:
            comparacoes += lista_temp[0]
            trocas += lista_temp[1]

    for i in range(n - 1, 0, -1):
        trocas += 1
        (lista[i], lista[0]) = (lista[0], lista[i])
        lista_temp = heapify(lista, i, 0)
        if lista_temp != None:
            comparacoes += lista_temp[0]
            trocas += lista_temp[1]
        
    after = datetime.now()
    return len(lista), comparacoes, trocas, (after - before).total_seconds()
```

```python
def dados_quick():
  dados = []
  dados.append(quickSort(dez.copy(), show_output=False))
  dados.append(quickSort(cem.copy(), show_output=False))
  dados.append(quickSort(mil.copy(), show_output=False))
  dados.append(quickSort(dezmil.copy(), show_output=False))
  dados.append(quickSort(cemmil.copy(), show_output=False))
  return dados
```

```python
def dados_shell():
  dados = []
  dados.append(shellSort(dez.copy()))
  dados.append(shellSort(cem.copy()))
  dados.append(shellSort(mil.copy()))
  dados.append(shellSort(dezmil.copy()))
  dados.append(shellSort(cemmil.copy()))
  return dados
```

```python
def dados_heap():
  dados = []
  dados.append(heapSort(dez.copy()))
  dados.append(heapSort(cem.copy()))
  dados.append(heapSort(mil.copy()))
  dados.append(heapSort(dezmil.copy()))
  dados.append(heapSort(cemmil.copy()))
  return dados
```

```python
def grafico(dados, title):
  colunas = ["Tamanho", "Comparações", "Trocas", "Tempo (s)"]
  df = pd.DataFrame(dados, columns=colunas)
  grafi, axes = plt.subplots(nrows = 1, ncols=3, figsize=(12, 4), tight_layout=True)
  grafi.suptitle(title, fontsize=18, fontweight='bold')
  
  for i in range(len(axes)):
    axes[i].plot(df[colunas[0]], df[colunas[i + 1]], color='red', linestyle='solid', markerfacecolor='blue', markersize=5) #, marker='o' - criar pontos
    axes[i].set_xlabel(colunas[0], fontsize='large', fontweight='bold')
    axes[i].set_ylabel(colunas[i+1], fontsize='large', fontweight='bold')
    axes[i].axhline(color="black")
    axes[i].axvline(color="black")
    axes[i].tick_params(axis='x', rotation=60)

  plt.show()
```

```python
def compara_tempo(a, b, c):
  plt.plot(a["Tamanho"], a["Tempo (s)"])
  plt.plot(b["Tamanho"], b["Tempo (s)"])
  plt.plot(c["Tamanho"], c["Tempo (s)"])
  plt.ylim(-5, c["Tempo (s)"].max())
  plt.legend(('QuickSort', 'HeapSort', 'ShellSort'),
           loc='upper right')
  plt.show()
```

```python
def dados_quic_0_100k(array_random):
  dados_quickSort = []
  for a in array_random:
    dados_quickSort.append(quickSort(a, show_output=False))
  return dados_quickSort
```

```python
def dados_heap_0_100k(array_random):
  dados_heap = []
  for a in array_random:
    dados_heap.append(heapSort(a))
  return dados_heap
```

```python
def dados_shell_0_100k(array_random):
  dados_shell = []
  for a in array_random:
    dados_shell.append(shellSort(a))
  return dados_shell
```

```python
def sai_graphics():
  start = []
  for k in range(0, 200_001, 1_000):
    a = (sample(range(0, 200_000), k))
    start.append(a)
  array_random_one = []
  array_random_two = []
  array_random_three = []
  for k in start:
    array_random_one.append(k.copy())
    array_random_two.append(k.copy())
    array_random_three.append(k.copy())
  a = pd.DataFrame(dados_quic_0_100k(array_random_one[:2]),columns=["Tamanho", "Comparações", "Trocas", "Tempo (s)"])
  grafico(a, "QuickSort")
  b = pd.DataFrame(dados_heap_0_100k(array_random_two[:2]),columns=["Tamanho", "Comparações", "Trocas", "Tempo (s)"])
  grafico(b, "HeapSort")
  c= pd.DataFrame(dados_shell_0_100k(array_random_three[:2]),columns=["Tamanho", "Comparações", "Trocas", "Tempo (s)"])
  grafico(c, "ShellSort")
  compara_tempo(a,b,c)
```
